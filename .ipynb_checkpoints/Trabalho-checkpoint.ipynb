```python
tcp = open("lbl-pkt-4.tcp.txt",'r')
l_tcp = tcp.readlines()
tcp.close()
for i in xrange(len(l_tcp)):
    l_tcp[i] = l_tcp[i].rstrip('\n')
    l_tcp[i] = l_tcp[i].split()
    for j in xrange(len(l_tcp[i])):
        if j == 0:
            l_tcp[i][j] = float(l_tcp[i][j])
        else:
            l_tcp[i][j] = int(l_tcp[i][j])
```
